Load training data

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

df_train = pd.read_csv("Datasets/poker-hand-training.csv")
X = df_train.iloc[:, :-1].values
y = df_train.iloc[:, -1].values
print("Training dataset ==> X.shape:", X.shape, "y.shape:", y.shape)
```

Visualize the training data 

```python
X_pca = PCA(n_components = 2).fit_transform(X)

print(f"shape of original data: {X.shape}, shape of transformed data: {X_pca.shape}")

# y value equals poker hand
X0_pca = X_pca[y==0]
X1_pca = X_pca[y==9]

# Visualize the data
plt.scatter(X0_pca[:, 0], X0_pca[:, 1], marker=".", color="green", label="Non-poker-hand (label 0)")
plt.scatter(X1_pca[:, 0], X1_pca[:, 1], marker="1", color="red", label="One Pair (label 1)")

plt.xlabel("First principal component of PCA")
plt.ylabel("Second principal component of PCA")
plt.title("Training dataset visualization using PCA")

plt.legend()
plt.show()
```

Train the model. SVC Linear model

```python
model = SVC(C=0.3, kernel="linear")
model.fit(X, y)
```

Load the test data

```python
# Loading the testing dataset into Xtest (inputs) and ytest (outputs)
df_test = pd.read_csv("Datasets/poker-hand-testing.csv")
Xtest = df_test.iloc[:1000, :-1].values # Take all columns except the last one
ytest = df_test.iloc[:1000, -1].values  # Take the last column (corresponding to the class-label)
print("Testing dataset ==> Xtest.shape:", Xtest.shape, "ytest.shape:", ytest.shape)


print(Xtest[0])
```

Test the model

```python

y_pred = model.predict(X)

ytest_pred = model.predict(Xtest)

acc_train = accuracy_score(y, y_pred)
acc_test = accuracy_score(ytest, ytest_pred)
print(f"Training Accuracy = {acc_train}, Test Accuracy = {acc_test}")
```
